```python
df = df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1)
```

```python
print("cleaned shape:", df.shape)
```

```python
attrition_count = df['Attrition'].value_counts()
print(attrition_count)

plt.figure(figsize=(6,4))
attrition_count.plot(kind='bar', color=['green','red'])
plt.title('Employee Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('attrition_overview.png')
plt.show()
print("16% employees left the company")
```

```python
dept_attrition = df[df['Attrition']=='Yes']['Department'].value_counts()
print(dept_attrition)

plt.figure(figsize=(6,4))
dept_attrition.plot(kind='bar', color='coral')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Employees Left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('dept_attrition.png')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set2')
plt.title('Monthly Income vs Attrition')
plt.tight_layout()
plt.savefig('salary_attrition.png')
plt.show()

avg_sal = df.groupby('Attrition')['MonthlyIncome'].mean()
print("Average Salary:")
print(avg_sal)
```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20, palette='Set1')
plt.title('Age Distribution vs Attrition')
plt.tight_layout()
plt.savefig('age_attrition.png')
plt.show()
```

```python
exp_attr = df.groupby('YearsAtCompany')['Attrition'].apply(
    lambda x: (x=='Yes').sum()
).reset_index()
exp_attr.columns = ['YearsAtCompany', 'AttritionCount']

plt.figure(figsize=(10,4))
sns.barplot(x='YearsAtCompany', y='AttritionCount', data=exp_attr, color='steelblue')
plt.title('Attrition by Years at Company')
plt.tight_layout()
plt.savefig('experience_attrition.png')
plt.show()
```

```python
import pymysql

conn = pymysql.connect(
    host='localhost',
    user='root',
    password='password',
    database='HR_project'
)

print("MySQL Connected !!")
```

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:password@localhost/HR_project')
df.to_sql('hr_data', engine, if_exists='replace', index=False)
print("Data Loaded into MySQL !!")
print("Total Records:", df.shape[0])
```

```python
cursor = conn.cursor()
cursor.execute("SHOW TABLES FROM HR_project;")
tables = cursor.fetchall()
print("Tables in your database:")
for table in tables:
    print(table)
```

```python
query1 = """
SELECT Attrition,
COUNT(*) as Total_Employees
FROM hr_data
GROUP BY Attrition;
"""

result1 = pd.read_sql_query(query1, conn)
print(result1)
```

```python
query2 = """
SELECT Department,
COUNT(*) as Employees_Left
FROM hr_data
WHERE Attrition = 'Yes'
GROUP BY Department
ORDER BY Employees_Left DESC;
"""

result2 = pd.read_sql_query(query2, conn)
print("=== DEPARTMENT WISE ATTRITION ===")
print(result2)
```

```python
query3 = """
SELECT Attrition,
ROUND(AVG(MonthlyIncome), 2) as Avg_Salary,
ROUND(MIN(MonthlyIncome), 2) as Min_Salary,
ROUND(MAX(MonthlyIncome), 2) as Max_Salary
FROM hr_data
GROUP BY Attrition;
"""

result3 = pd.read_sql_query(query3, conn)
print("=== SALARY VS ATTRITION ===")
print(result3)
```

```python
query4 = """
SELECT 
CASE 
    WHEN Age < 25 THEN 'Below 25'
    WHEN Age BETWEEN 25 AND 35 THEN '25-35'
    WHEN Age BETWEEN 36 AND 45 THEN '36-45'
    ELSE 'Above 45'
END as Age_Group,
COUNT(*) as Employees_Left
FROM hr_data
WHERE Attrition = 'Yes'
GROUP BY Age_Group
ORDER BY Employees_Left DESC;
"""

result4 = pd.read_sql_query(query4, conn)
print("=== AGE GROUP ATTRITION ===")
print(result4)
```

```python
query5 = """
SELECT 
CASE 
    WHEN YearsAtCompany < 2 THEN '0-2 Years'
    WHEN YearsAtCompany BETWEEN 2 AND 5 THEN '2-5 Years'
    WHEN YearsAtCompany BETWEEN 6 AND 10 THEN '6-10 Years'
    ELSE 'Above 10 Years'
END as Experience_Group,
COUNT(*) as Employees_Left
FROM hr_data
WHERE Attrition = 'Yes'
GROUP BY Experience_Group
ORDER BY Employees_Left DESC;
"""

result5 = pd.read_sql_query(query5, conn)
print("=== EXPERIENCE VS ATTRITION ===")
print(result5)
```
